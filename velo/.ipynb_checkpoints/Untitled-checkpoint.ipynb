```python
def renameLift(x):
    if x.lower() == 'bp':
        return 'bench'
    elif x.lower() == 'floorp':
        return 'floor'
    elif x.lower() == 'fsq':
        return 'front squat'
    elif x.lower() == 'bsq':
        return 'squat'
    elif x.lower() == 'goodmrng':
        return 'gm'
    elif x.lower() == 'dl':
        return 'dead'
    elif x.lower() == 'bndbp':
        return 'band bench'
    elif x.lower() in ['sdl','sumo','sumo deadlift']:
        return 'sumo'
    elif x.lower() == 'pausebsqt':
        return 'pause squat'
    elif x.lower() == 'bndbsq':
        return 'band squat'
    elif x.lower() == 'boardpr':
        return 'board'
    else:
        return x.lower()
```

```python
def getLog(sheetName):
    sheet_id='1TlhsIhpwUeQan4o_97tfNeeztTJinQTQvLByTj5YF9s'
    sheet_name = sheetName

    url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
    train = pd.read_csv(url)

    log = pd.DataFrame()
    for i in range(6):
        temp = train.iloc[:,np.r_[:2,2+6*i:8+6*i]]
        temp.columns=[a[:a.find('.')].lower() if a.find('.')>-1 else a.lower() for a in temp.columns]
        temp.rename(columns={'week':'tm','unnamed: 1':'type','sets\nxreps':'sets_reps','last set\ngoal':'last_set_goal',
                            'last\nset':'last_set'},inplace=True)
        temp['wk'] = i+1
        log = log.append(temp)
    log.reset_index(drop=True,inplace=True)
    return log
```

```python
def importLoad(path = '/Users/ericcoxon/Library/Mobile Documents/com~apple~CloudDocs/Training/'):
    files = [path+f for f in os.listdir(path) if f.find('.csv')>-1 and f.find('T000')==-1 ]
    df = pd.DataFrame()
    for file in files:
        print('going through:',file)
        name = file.replace(path,'')
        temp = pd.read_csv(file,skiprows=3)
        temp.Exercise = renameLift(name[:name.find('-')])
        #temp.Date = file[file.find('-'):file.find('.csv')]
        temp =temp.fillna(method='ffill')
        temp.dropna(axis=0,subset=['Type*'],inplace=True)
        temp.Date = pd.to_datetime(temp.Date)
        temp['day'] = temp.Date.dt.strftime('%a')
        temp['time'] = temp.Date.dt.time
        if len([col for col in temp.columns if col.find('kg')>-1])>0:
            temp['wgt_lb'] = temp['Weight (kg)'] * 2.2
            temp['rom_ft'] = temp['ROM (cm)'] / 2.54 / 12
        df = df.append(temp)
    df.rename(columns={'# Set':'set','# Rep.':'rep','Type*':'ecc_con','MPV (m/s)':'mpv_ms','Peak Velocity (m/s)':'peak_velo',
                       'Mean Power (W)':'mean_watts','ROM (cm)':'rom_cm','1RM (kg)':'est_max_kg','Weight (kg)':'wgt_kg',
                      'Exercise':'lift','Date':'date','Weight (lb)':'wgt_lb','1RM (lb)':'est_max_lb','ROM (ft)':'rom_ft'},inplace=True)
    return df
```

```python
df = pd.DataFrame()
for a in files:
    temp = pd.read_csv(a,sep='\t',skiprows=3)
    temp.Date = temp.Date[0]
    temp.Exercise = temp.Exercise[0].title().strip()
    #temp['file'] = a[a.find('/Training')+10:a.find('-')].strip()
    temp.Date = pd.to_datetime(temp.Date).dt.date
    temp['day'] = pd.to_datetime(temp.Date[0]).strftime('%A')
    df = df.append(temp)
df.rename(columns={'# Set':'set','# Rep.':'rep','Type*':'ecc_con','MPV (m/s)':'mpv_ms','Peak Velocity (m/s)':'peak_velo',
                   'Mean Power (W)':'mean_watts','ROM (ft)':'rom_ft','1RM (lb)':'est_max','Weight (lb)':'wgt_lb',
                  'Exercise':'lift','Date':'date'},inplace=True)
df.date = pd.to_datetime(df.date)
df = df.dropna(thresh=4)
df = df.mask(df.set.notna(),df.fillna(0)).fillna(method='ffill')
df = df.loc[(df.ecc_con!=0) & (df.ecc_con.notna())]
#df.lift = df.lift.mask((df.file!='DataRecords')&(~df.file.isin(df.lift)),df.file)
df.lift = df.lift.mask(df.lift.str.lower() == 'lowbar back squat','Squat')
df.lift = df.lift.mask(df.lift.str.lower() == 'high bar squat','Ssb')
```
